```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sys
import matplotlib
from random import choice
%matplotlib inline

class RandomWalk:
    def __init__(self, num_pts=5000):
        self.num_pts=num_pts
        self.x_val = [0]
        self.y_val = [0]
    
    def walk(self):
        while len(self.x_val)<self.num_pts:
            x_dir = choice([-1, 1])
            x_dist = choice([0, 1, 2, 3, 4, 5])
            y_dir = choice([-1, 1])
            y_dist = choice([0, 1, 2, 3, 4, 5])
            
            x_step = x_dir*x_dist
            y_step = y_dir*y_dist
            if x_step==0 and y_step==0:
                continue
            x_next = self.x_val[-1] + x_step
            y_next = self.y_val[-1] + y_step
            self.x_val.append(x_next)
            self.y_val.append(y_next)

rw = RandomWalk()
rw.walk()
#plt.scatter(rw.x_val, rw.y_val, s=10)
pts_idx = list(range(rw.num_pts))
plt.figure(figsize=(15,10))
plt.scatter(rw.x_val, rw.y_val, c=pts_idx, s=10)
#plt.scatter(rw.x_val, rw.y_val, c=pts_idx, cmap=plt.cm.Reds, s=10)
# start point
plt.scatter(rw.x_val[0], rw.y_val[0], c='red')
plt.scatter(rw.x_val[-1], rw.y_val[-1], c='blue')
plt.axes().get_xaxis().set_visible(False)
plt.axes().get_yaxis().set_visible(False)
#plt.show()
plt.savefig('randomwalk.png')
```
